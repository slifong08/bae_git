```r
library(ggplot2)
```

```r
output_path = "/dors/capra_lab/users/fongsl/broadly_active_enhancers/data/roadmap_multi_data/"

#The original multi-intersected roadmap data
data = "short_multiintersect_roadmap.bed"

#The multi-intersected roadmap data 99% of overlapping samples.
data99 = "99_short_multiintersect_roadmap.bed"
```

```r
multi99 = read.csv(paste0(output_path, data99), sep = '\t', header = F)
```

```r
multi = read.csv(paste0(output_path, data), sep = '\t', header =F)
```

```r
head(multi99)
dim(multi99)
```

```r
# format the dataframes with column names
names = c("chr", "start", "end", "num", "sample_id","length")
multi["length"] = multi$end - multi$start
colnames(multi99) = names
```

```r
ggplot(multi, aes(x =num, fill = factor(num))) +
  geom_histogram(bins = 70)+
    scale_x_continuous(name = "Roadmap Samples")+
    ggtitle("Roadmap Enhancers Overlapping Tissues\n all")+
    labs(x = "Roadmap Samples", y= "No.Enhancers")
```

```r
ggplot(multi99, aes(x =num, y =length, fill = factor(num))) +
  geom_boxplot()+
    scale_x_continuous(name = "Roadmap Samples")+
    ggtitle("Roadmap Enhancers Overlapping Tissues \n99th percentile")+
    labs(x = "Roadmap Samples", y= "Length of Enhancer")
```

```r
ggplot(multi99, aes(x =num, fill = factor(num))) +
  geom_histogram(bins = 33)+
    scale_x_continuous(name = "Roadmap Samples")+
    ggtitle("Human Specific Overlapping Tissues \n 99th percentile")+
    labs(x = "Roadmap Samples", y= "No.Enhancers *fragments*")
```
