```r
library(ggplot2)
```

```r
output_path = "/dors/capra_lab/users/fongsl/broadly_active_enhancers/baes_data/roadmap_multi_data/"

#The original multi-intersected roadmap data
data = "short_multiintersect_roadmap.bed"

#The multi-intersected roadmap data NOT overlapping other species by subtraction.
pcpl = "intersection_pcpl.bed"
pcpr = "intersection_pcpr.bed"
pcve = "intersection_pcve.bed"
```

```r
pl = read.table(paste0(output_path, pcpl),sep = '\t')
pr = read.table(paste0(output_path, pcpr),sep = '\t')
ve = read.table(paste0(output_path, pcve),sep = '\t')
```

```r
multi = read.csv(paste0(output_path, data), sep = '\t')
```

```r
head(multi)
```

```r
# format the dataframes with column names, peak lengths
names = c("chr", "start", "end", "num", "sample_id")
colnames(multi) = names
colnames(pl) = names
colnames(pr) = names
colnames(ve) = names

pl$length = pl$end - pl$start
pr$length = pr$end - pr$start
ve$length = ve$end - ve$start
dim(pl)
```

```r
#density plot
pl.dl<-density(pl$length)
pr.dl<-density(pr$length)
ve.dl<-density(ve$length)

par = c(0.5,0.5, 0.5,0.5)
plot(pl.dl, main = "Human Specific Peak Length", xlab = "peak length")
lines(pr.dl)
lines(ve.dl)
```

```r
hist(pl$length, breaks = 50, col=rgb(1,0,0,0.25), main = "Human Specific Peak Length", lend=2, ylim =c(0,7000000), xlab="Peak Length")
hist(pr$length, breaks = 50, col=rgb(0,1,0,0.25), add=T,lend=2)
hist(ve$length, breaks = 50, col=rgb(0,0,1,0.25), add=T,lend=2)
legend("topright", inset=.05, title="Substracted PhastCon",
  	c("Pl","Pr","Ve"), fill=c(rgb(1,0,0,0.25), rgb(0,1,0,0.25), rgb(0,0,1,0.25)), horiz=TRUE)
```

```r
ggplot(multi, aes(x =num, fill = factor(num))) +
  geom_histogram(bins = 70)+
    scale_x_continuous(name = "Roadmap Samples")+
    ggtitle("Roadmap Enhancers Overlapping Tissues \nPre-PhastCon")+
    labs(x = "Roadmap Samples", y= "No.Enhancers")
```

```r
ggplot(multi, aes(x =percentile, y =length, fill = factor(num))) +
  geom_boxplot()+
    scale_x_continuous(name = "Roadmap Samples")+
    ggtitle("Roadmap Enhancers Overlapping Tissues \nPre-PhastCon")+
    labs(x = "Roadmap Samples", y= "Length of Enhancer")
```
